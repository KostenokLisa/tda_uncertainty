## Statistics from barcodes

```python
features_ripser = np.load("/content/sample_data/train_ripser.npy")
features_matrix_0_0 = features_ripser[0, 0, :, :]
```

```python
features_matrix_0_0.shape
```

```python
features_df_0_0 = pd.DataFrame(features_matrix_0_0)
features_df_0_0.head()
```

```python
ripser_feature_names=[
    'h0_s', 
    'h0_e',
    'h0_t_d', 
    'h0_n_d_m_t0.75',
    'h0_n_d_m_t0.5',
    'h0_n_d_l_t0.25',
    'h1_t_b',
    'h1_n_b_m_t0.25',
    'h1_n_b_l_t0.95', 
    'h1_n_b_l_t0.70',  
    'h1_s',
    'h1_e',
    'h1_v',
    'h1_nb'
]
```

```python
features_df_0_0.columns = ripser_feature_names
```

```python
features_df_0_0.hist(figsize=(40,30)) # 0 layer 0 head
```

```python
# 1 layer 0 head
features_matrix_1_0 = features_ripser[1, 0, :, :]
features_df_1_0 = pd.DataFrame(features_matrix_1_0)
features_df_1_0.columns = ripser_feature_names
features_df_1_0.hist(figsize=(40,30))
```

```python
# 11 layer 0 head
features_matrix_11_0 = features_ripser[11, 0, :, :]
features_df_11_0 = pd.DataFrame(features_matrix_11_0)
features_df_11_0.columns = ripser_feature_names
features_df_11_0.hist(figsize=(40,30))
```

```python
# 0 layer 11 head
features_matrix_0_11 = features_ripser[0, 11, :, :]
features_df_0_11 = pd.DataFrame(features_matrix_0_11)
features_df_0_11.columns = ripser_feature_names
features_df_0_11.hist(figsize=(40,30))
```

```python
for i, feature in enumerate(ripser_feature_names, 0):
    features_mean = np.mean(features_ripser[:, :, :, i])
    features_var = np.var(features_ripser[:, :, :, i])
    features_median = np.median(features_ripser[:, :, :, i])
    print("feature", i)
    print("mean =", round(features_mean, 3), "median =", round(features_median, 3))
    print("var =", round(features_var, 3))
    print("eps =", round(np.sqrt(features_var) / features_mean, 3))
```
